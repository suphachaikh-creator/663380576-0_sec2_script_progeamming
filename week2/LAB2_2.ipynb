```python
# 1. รับอายุจากผู้ใช้ด้วย input() และแปลงเป็นจำนวนเต็มด้วย int()
age = input("กรุณากรอกอายุของคุณ: ")

# รับค่า like_action (yes/no)
like_action = input("คุณชอบหนังแอ็กชันไหม? (yes/no): ")

if age.isdigit() and like_action == "yes" or like_action =="no":#ตรวจว่าค่าแปรเป็นตัวเลขมั้ย และ กรอก yes หรือ no ถูกมั้ย

  age = int(age) #แปลงค่าเป็นจำนวนเต็ม

  if age >= 18 and like_action == "yes": #หากมากกว่าหรือเท่ากับ 18 และชอบแอ็กชัน
      print("หนังแอ็กชันเรต R")#แสดง หนังแอ็กชันเรต R

  elif age >= 18 and like_action == "no":#หากมากกว่าหรือเท่ากับ 18 และไม่ชอบแอ็กชัน
      print("หนังดราม่า/ไซไฟ")#แสดง หนังดราม่า/ไซไฟ

  elif age >= 13 and like_action == "yes": #หากมากกว่าหรือเท่ากับ 13 และชอบแอ็กชัน
      print("หนังแอ็กชันฮีโร่เรต PG-13")#แสดง หนังแอ็กชันฮีโร่เรต PG-13

  elif age >= 13 and like_action == "no": #หากมากกว่าหรือเท่ากับ 13 และไม่ชอบแอ็กชัน
      print("หนังผจญภัยทั่วไป")#แสดง หนังผจญภัยทั่วไป

  elif age < 13 and like_action == "yes": #หากน้อยกว่า 13 และชอบแอ็กชัน
      print("หนังแอนิเมชันทั่วไป")#แสดง หนังแอนิเมชันทั่วไป

  else:  #ไม่ตรงตามข้อไหนเลย
      print("ภาพยนตร์ทั่วไปสำหรับครอบครัวหรือเด็ก")#แสดง ภาพยนตร์ทั่วไปสำหรับครอบครัวหรือเด็ก
else:
  print("=== กรอก yes/no ไม่ถูก หรือ กรอกตัวเลขผิด ===") #แสดง === กรอก yes/no ไม่ถูก หรือ กรอกตัวเลขผิด ===
```
