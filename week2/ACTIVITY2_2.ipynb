```python
import random

choices = ["ค้อน", "กระดาษ", "กรรไกร"]

# 🎯 [แก้ไขจุดที่ 2] ย้ายตัวแปรเก็บคะแนนมาไว้นอก while loop เพื่อไม่ให้คะแนนรีเซ็ตทุกตา
num = 0  # คะแนนของคุณ
com = 0  # คะแนนของคอม

while True: #ลูปเรื่อยๆ จนกว่าได้รับคำสั่ง break
    player = input("\nออกอะไร? (ค้อน/กระดาษ/กรรไกร) หรือพิมพ์ 'ออก': ")

    if player == "ออก": #ตัวแปร == ออก จะเข้าเงื่อนไข
        print("บ๊ายบาย!")
        break#ออกจากลูป

    if player not in choices:#ตัวแปร ไม่ตรงตามค่าที่อยู่ใน choices
        print("พิมพ์ไม่ถูกต้อง เอาใหม่นะ!")
        continue #เริ่มต้นลูปใหม่

    bot = random.choice(choices) #คำสั่ง random.choice() สุ่มจาก choices
    print(f"คุณ: {player} VS คอม: {bot}") #แสดงการออกค่า

    #กรณีเสมอ
    if player == bot:#ค่าตรงกัน
        print("🤝 เสมอกัน!")

    #กรณีคุณชนะ
    elif (player == "ค้อน" and bot == "กรรไกร") or \
         (player == "กระดาษ" and bot == "ค้อน") or \
         (player == "กรรไกร" and bot == "กระดาษ"): #ใส่กรณีชนะทั้งหมดของผู้เล่น
        print("🎉 คุณชนะ!")
        num += 1  #บวก 1

        #หาก num เท่ากับ ตัวเลขตามกำหนด แสดงบทสนทนา
        if num == 1:
            print("คอม: ว้าว เก่งจังแต่ฉันไม่แพ้ง่ายๆหรอกนะ")
        elif num == 2:
            print("คอม: ฉันจะแพ้แล้วว ไม่นะะ")
        elif num == 3:
            print("คอม: แพ้แล้ว TT")

        #แสดงคะแนนทั้งสอง
        print(f"สกอร์ตอนนี้ -> คุณ {num} / คอม {com}")

        #หากครบชนะครบ 3 แสดงข้อความยินดี แล้วจบเกม
        if num == 3:
            print("\n🏆 ยินดีด้วยคุณชนะทัวร์นาเมนต์นี้!")
            break

    #กรณีคุณแพ้
    else:
        print("❌ คุณแพ้!")
        com += 1  #con บวก 1
        print(f"สกอร์ตอนนี้ -> คุณ {num} / คอม {com}")

        #ถ้าคอมได้ 3 คะแนน คอมจะเยาะเย้ยและจบเกม
        if com == 3:
            print("\nจบเกม")# \n คือเว้นช่อง
            print("คอม: EZ 😎 (คุณแพ้ราบคาบ)")
            break
```
