```python
num = int(input("Enter a number: ") )
if num < 0:
    text = f"{num} is negative"
elif num > 0:
    text = f"{num} is positive"
else:
    text = f"{num} is zero"

if num == 0:
    print(f"{text}.")
elif num % 2 == 0:
    print(f"{text} and even number.")
else:
    print(f"{text} and odd number.")
```

ตรวจจำนวนเต็มบวก,เต็มลบ,ศูนย์ ด้วย if-elif-else โดยหาค่าด้วย ศูนย์ หากมากกว่าจะเป็นเต็มบวก หากน้อยกว่าเป็นเต็มลบ หากเท่ากับเป็นศูนย์
