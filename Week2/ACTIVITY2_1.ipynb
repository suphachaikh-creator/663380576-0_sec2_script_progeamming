```python

# Block A: Evaluation Check (การตรวจสอบเงื่อนไขสภาพอากาศ)
temp = 25          # กำหนดค่าอุณหภูมิ
is_raining = True  # กำหนดสถานะฝนตก (True = ฝนตก, False = ฝนไม่ตก)


# temp > 20 (25 มากกว่า 20 = True) AND not is_raining (ตรงข้ามกับ True = False)
if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")  # จะพิมพ์ก็ต่อเมื่อ อุณหภูมิมากกว่า 20 และฝนต้องไม่ตก

# temp <= 20 (25 น้อยกว่าหรือเท่ากับ 20 = False) OR is_raining (ค่าคือ True)
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")

#กรณีที่ไม่เข้าพวกกับเงื่อนไขด้านบนเลย
else:
    print("Unsure about the weather.")


# Block B: Vectorized Filter
phrase = "Python 2026 Engine"

# 2. ใช้ Set Comprehension ตัวกรองในบรรทัดเดียวซ้อน Loop ลุยอ่านทีละตัวอักษร (c)
# - c.lower()               : แปลงตัวอักษรที่ผ่านการกรองให้เป็นตัวพิมพ์เล็กก่อนเก็บเข้า Set
# - for c in phrase         : วนลูปอ่านตัวอักษรทีละตัวจากข้อความ "Python 2026 Engine"
# - if c.isalpha()          :ต้องเป็นตัวอักษรเท่านั้น (ตัวเลขและช่องว่างโดนคัดออก)
# - and c.lower() not in 'aeiou' :ตัวพิมพ์เล็กของมันต้องไม่ใช่สระ (a, e, i, o, u)
consonants = {c.lower() for c in phrase if c.isalpha() and c.lower() not in 'aeiou'}

#แสดงผลลัพธ์
print(f"Unique consonants: {consonants}")
```
